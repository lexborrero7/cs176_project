```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
full_df = pd.read_csv("Advanced.csv", sep=",")
college_df = pd.read_csv("nba_college.csv", sep=",")
player_df = pd.read_csv("all_seasons.csv", sep=",")
```

```python
# filter so that only specific columns are used
player_df = player_df[["player_name", "team_abbreviation", "age", "player_height", "player_weight", "college", "country", "draft_year", "draft_round", "gp", "pts", "reb", "ast"]]
display(player_df)
```

```python
# drop missing values by deleting the column because it exists in another dataset
player_df = player_df.drop("college", axis=1)
```

```python
# remove duplicates
duplicated = player_df.drop_duplicates()
```

```python
# subset the dataset into various dataframes

# filter to get only USA players
usa_players = player_df[(player_df["country"] == "USA")]
display(usa_players)

# filter to get players that have more than 20 points per game
high_scorers = player_df[(player_df["pts"] > 20)]
display(high_scorers)
```

Indicates that 10,721 rows out of 12,844 rows are USA athletes.

Also shows that out of 12,844 rows, only 701 are top scorers with over 20 points per game.

```python
# sort data in various ways

# sort so that it decreases by age first and then increases by points second
age_pts_df = player_df.sort_values(by=['age', 'pts'], ascending=[False, True])
display(age_pts_df)

# sort the subsetted dataframe for high scorers by average rebounds and assists per game
high_scorers_reb = high_scorers.sort_values(by=["reb", "ast"], ascending=False)
display(high_scorers_reb)
```

The first sort helps to identify the players that are old yet high scorers. It also helps identify the players who are younger and still had a lot of points.

The second sort helps visualize who the top scorers mixed with the most rebounds and assists per game are. Essentially, telling us who valuable players on teams were (MVP).
